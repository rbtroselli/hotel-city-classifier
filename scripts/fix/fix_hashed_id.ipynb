```python
import hashlib
import sqlite3

def _get_hashed_id(string):
    # Calculate the SHA-256 hash of the string
    hash_value = hashlib.sha256(string.encode()).hexdigest()
    # Convert the hexadecimal hash value to an integer
    hash_int = int(hash_value, 16)
    # Truncate the integer to 20 digits. It has sufficient collision resistance for this use case
    truncated_hash = hash_int % (10 ** 18)
    return truncated_hash

connection = sqlite3.connect('/Users/robertoroselli/Code/hotel-city-classifier/hotel.db')
cursor = connection.cursor()

cursor.execute('SELECT * FROM HOTEL_URL_HOTEL_ID_BRIDGE')
rows = cursor.fetchall()
for row in rows:
    hotel_id = row[1]
    url = row[0]
    hashed_id = _get_hashed_id(url)
    print(hotel_id, url, hashed_id)
    cursor.execute('UPDATE HOTEL_URL_HOTEL_ID_BRIDGE SET HOTEL_ID_NEW = ? WHERE HOTEL_ID = ?', (hashed_id, hotel_id))

connection.commit()
# # Example usage
# string = "ciao"
# hashed_value = _get_hash_from_string(string)
# print(hashed_value)
```
